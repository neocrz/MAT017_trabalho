```python
# Função que calcula o gradiente da função em (x, y)
def grad(x,y):
  return[x**23,2*y]
```

```python
import math
# Função que calcula a distância entre dois pontos
def dist(anterior,novo):
  zs = zip(anterior,novo)
  arc = 0
  for[p1,p2] in zs:
    arc = arc + (p1-p2)**2
  return math.sqrt(arc)
```

```python
# Função que implementa a descida de gradiente
def grad_desc(lr,xn,yn,tol):
  xn1 = 999999
  yn1 = 999999
  k = 0
  d = tol +  1
  # Listas para armazenar os pontos visitados
  x_points = [xn]
  y_points = [yn]
  while d > tol: # Continua até a distância ser menor que a tolerância
    # Calcula novos pontos baseados no gradiente e na taxa de aprendizado
    xn1 = xn - lr * grad(xn,yn)[0]
    yn1 = yn - lr * grad(xn,yn)[1]
    # Calcula a distância entre os pontos antigos e novos
    d = dist([xn, yn], [xn1, yn1])
    k = k + 1
    # Atualiza os pontos atuais
    xn = xn1
    yn = yn1
    # Adiciona os novos pontos às listas
    x_points.append(xn)
    y_points.append(yn)
  # Retorna as coordenadas finais, o número de iterações e os pontos visitados
  return [xn1,yn1,k,x_points, y_points]
```

```python
# Executa a descida de gradiente
z = grad_desc(0.01,1,1,1e-4)
z
```

```python
import matplotlib.pyplot as plt
# Recupera os pontos da trajetória
x_points = z[3]
y_points = z[4]

# Cria o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x_points, y_points, marker='o', linestyle='-')
plt.title('Trajetória da Descida de Gradiente')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
